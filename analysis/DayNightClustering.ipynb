```python
import pandas as pd
import numpy as np

import time
import datetime

DATASETS_LOCATION = "/Users/gomerudo/workspace/git_storage/tuedataeng/visualization/results"
VARIABLE = 'Location'
# VARIABLE = 'Crime'
N_CLUSTERS = 3
DATASET_CRIMES_PATH = DATASETS_LOCATION + "/chicagoAllLocations.csv"
# DATASET_CRIMES_PATH = DATASETS_LOCATION + "/chicagoAllTrueCrimes.csv"
```

```python
# Loading from absolute path, because the dataset is too big to download and to upload

# DATASET_CRIMES_PATH = DATASETS_LOCATION + "/chicagoAllTrueCrimes.csv"



crimes_df = pd.read_csv(DATASET_CRIMES_PATH, na_values = ["", " "])
crimes_df = crimes_df.sort_values(['Community', 'Period', 'Frequency'])
crimes_df
```

```python
crimes_day_df = crimes_df[crimes_df['Period'] == 'Day']
crimes_day_df
```

```python
crimes_night_df = crimes_df[crimes_df['Period'] == 'Night']
crimes_night_df
```

```python
locations = crimes_df[VARIABLE].unique()
locations = np.sort(locations)
locations

communities = crimes_df['Community'].unique()
communities = np.sort(communities)
communities

day_results = []

# For each community 
for community in communities:
    # Get the locations of those communities
    current_community_df = crimes_day_df[crimes_day_df['Community'] == community]
    # Create its entry full of zeros
    new_entry = np.zeros((len(locations),), dtype=int)
    # Iterate all crimes for that community
    for index, row in current_community_df.iterrows():
        # Get the location index in our array
        location_index = np.argwhere(locations == row[VARIABLE])[0][0]
        # Set that location index to the frequency
        new_entry[location_index] = row['Frequency']
    # Once 
    day_results.append(new_entry)

from sklearn.cluster import KMeans 
from sklearn.metrics import confusion_matrix

# N_CLUSTERS = 5
kmeans = KMeans(n_clusters = N_CLUSTERS, random_state = 0).fit(day_results)

kmeans.labels_
results_day = communities.copy().T
results_day = np.vstack([results_day, np.array(len(communities)*['Day'])])
results_day = np.vstack([results_day, kmeans.labels_]).T
# results_day = pd.DataFrame(results_day, columns=['Community', 'Period', 'Cluster'])
# results_day
```

```python
locations = crimes_df[VARIABLE].unique()
locations = np.sort(locations)
locations

communities = crimes_df['Community'].unique()
communities = np.sort(communities)
communities

night_results = []

# For each community 
for community in communities:
    # Get the locations of those communities
    current_community_df = crimes_night_df[crimes_night_df['Community'] == community]
    # Create its entry full of zeros
    new_entry = np.zeros((len(locations),), dtype=int)
    # Iterate all crimes for that community
    for index, row in current_community_df.iterrows():
        # Get the location index in our array
        location_index = np.argwhere(locations == row[VARIABLE])[0][0]
        # Set that location index to the frequency
        new_entry[location_index] = row['Frequency']
    # Once 
    night_results.append(new_entry)

from sklearn.cluster import KMeans 
from sklearn.metrics import confusion_matrix

# N_CLUSTERS = 7
kmeans = KMeans(n_clusters = N_CLUSTERS, random_state = 0).fit(night_results)

kmeans.labels_
results_night = communities.copy().T
# results_day.append(kmeans.labels_)
results_night = np.vstack([results_night, np.array(len(communities)*['Night'])])
results_night = np.vstack([results_night, kmeans.labels_]).T
# results_night = pd.DataFrame(results_night, columns=['Community', 'Period', 'Cluster'])
# results_night
```

```python
results_all = np.vstack([results_day, results_night])
results_all = pd.DataFrame(results_all, columns=['Community', 'Period', 'Cluster'])
results_all
```

```python
CLUSTERS_PERIOD_LOCATION_PATH = DATASETS_LOCATION + "/clusters_{}.csv".format(VARIABLE)

results_all.to_csv(CLUSTERS_PERIOD_LOCATION_PATH, index = False)
```
